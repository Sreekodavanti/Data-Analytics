```python
import requests
from bs4 import BeautifulSoup

# Function to scrape quotes from a given page
def scrape_quotes(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Extracting quotes
    quotes = soup.find_all('div', class_='quoteText')
    for quote in quotes:
        text = quote.text.strip().split('\n')[0]  # Extracting only the text part of the quote
        print(text)
    print('\n---\n')

# Scraping quotes from the first three pages of Goodreads Quotes
base_url = 'https://www.goodreads.com/quotes?page='

for page_number in range(1, 4):
    url = base_url + str(page_number)
    scrape_quotes(url)
```

strip()Method: This method is commonly used when processing user inputs, file reading, or web scraping to clean up strings and make them more manageable or presentable. In the context of web scraping, it's often used to clean up the text content extracted from HTML elements.

```python
original_string = "   Hello, World!   "
stripped_string = original_string.strip()

print("Original String:", original_string)
print("Stripped String:", stripped_string)
```
